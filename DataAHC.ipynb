```python
# import library
from matplotlib import pyplot as plt 
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel("Data AHC.xlsx")
print(data)
```

```python
x = np.asarray(data)
print(x)
```

```python
# Lakukan Hierarichal Klastering
Z = linkage(x)
print(Z)
```

```python
# plot dendogram
plt.figure(figsize=(25,10))
plt.title('Dendrogram Gierarichal Klastering')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(Z)
plt.show()
```
